```python
cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS',
        'RAD','TAX','PTRATIO','B','LSTAT','MEDV']
```

```python
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

# filter data (keep only non-outliers)
df_clean = df[~((df[cols] < (Q1 - 1.5 * IQR)) |
                (df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("Original shape:", df.shape)
print("After removing outliers:", df_clean.shape)
```

```python
df_clean
```

```python
X = df_clean.drop('MEDV', axis=1)   # features
y = df_clean['MEDV']                # target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# -------------------------------
# 1. Train-Test Split FIRST
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
# -------------------------------
# 2. Scaling (fit ONLY on train)
# -------------------------------
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1)).ravel()
```

```python
# -------------------------------
# 3. Model Training
# -------------------------------
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
# -------------------------------
# 4. Predictions (scaled space)
# -------------------------------
y_train_pred_scaled = model.predict(X_train_scaled)
y_test_pred_scaled = model.predict(X_test_scaled)
```

```python
# -------------------------------
# 5. RMSE (scaled - recommended for comparison)
# -------------------------------
train_rmse_scaled = np.sqrt(mean_squared_error(y_train_scaled, y_train_pred_scaled))
test_rmse_scaled = np.sqrt(mean_squared_error(y_test_scaled, y_test_pred_scaled))

print("Train RMSE (scaled):", train_rmse_scaled)
print("Test RMSE (scaled):", test_rmse_scaled)
```

```python
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print(coefficients.sort_values(by='Coefficient', ascending=False))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(y_test_scaled, y_test_pred_scaled)
plt.xlabel("Actual Values (scaled)")
plt.ylabel("Predicted Values (scaled)")
plt.title("Actual vs Predicted")
plt.plot([-3, 3], [-3, 3], color='red')  # reference line
plt.show()
```

```python
residuals = y_test_scaled - y_test_pred_scaled

plt.figure(figsize=(6,4))
plt.scatter(y_test_pred_scaled, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted Values (scaled)")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.kdeplot(residuals, fill=True)
plt.xlabel("Residuals")
plt.ylabel("Density")
plt.title("KDE Plot of Residuals")
plt.show()
```
